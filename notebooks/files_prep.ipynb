```python
import os
import pandas as pd

train_dir = r"C:\Users\abdoe\Downloads\aclImdb\train"
test_dir = r"C:\Users\abdoe\Downloads\aclImdb\test"

def load_reviews_from_folder(folder, label):
    reviews = []
    for filename in os.listdir(folder):
        if filename.endswith(".txt"):
            with open(os.path.join(folder, filename), "r", encoding="utf-8") as f:
                text = f.read()
                reviews.append((text, label))
    return reviews

def load_dataset(base_path):
    pos_reviews = load_reviews_from_folder(os.path.join(base_path, "pos"), 1)
    neg_reviews = load_reviews_from_folder(os.path.join(base_path, "neg"), 0)
    all_reviews = pos_reviews + neg_reviews 
    return pd.DataFrame(all_reviews, columns=["review", "label"])

train_df = load_dataset(train_dir)
test_df = load_dataset(test_dir)

print("Train Shape:", train_df.shape)
print("Test Shape:", test_df.shape)

train_df.to_csv("train_reviews.csv", index=False)
test_df.to_csv("test_reviews.csv", index=False)
```
